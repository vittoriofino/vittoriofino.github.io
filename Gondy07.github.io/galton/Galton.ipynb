# Previsione delle altezze future dei figli

Le analisi sono state svolte sui dati dell'altezza dei genitori e dei figli (training set Galton.txt)

## Grafico lineare tra l'altezza dei figli e la media dell'altezza dei genitori

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Funzione per convertire pollici in centimetri
def inches_to_cm(inches):
    return inches * 2.54

# Leggi il file Galton.txt
data = pd.read_csv('Galton.txt', sep='\t')

# Converti le altezze da pollici a centimetri
data['Father_cm'] = inches_to_cm(data['Father'])
data['Mother_cm'] = inches_to_cm(data['Mother'])
data['Height_cm'] = inches_to_cm(data['Height'])

# Calcola l'altezza media dei genitori
data['Average_Parent_Height'] = (data['Father_cm'] + data['Mother_cm']) / 2

# Seleziona le colonne per la regressione
X = data[['Average_Parent_Height']]
y = data['Height_cm']

# Suddivisione dei dati in set di addestramento e test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# Crea il modello di regressione lineare
model = LinearRegression()
model.fit(X_train, y_train)

# Predici le altezze
predictions = model.predict(X_test)

# Calcola il punteggio di accuratezza
score = model.score(X_test, y_test)

# Crea il grafico
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Dati reali')
plt.plot(X_test, predictions, color='red', linewidth=2, label='Regressione lineare')
plt.title('Regressione Lineare dell\'Altezza dei Figli')
plt.xlabel('Altezza Media dei Genitori (cm)')
plt.ylabel('Altezza dei Figli (cm)')
plt.legend()
plt.grid()
plt.show()

# Stampa il punteggio di accuratezza
print(f"Punteggio di accuratezza del modello: {score:.2f}")
```

Da questo grafico possiamo notare che nel tempo l'altezza dei figli è sempre aumentata e la media delle altezze attuali è i 180cm

## Grafici lineare tra l'altezza dei figli e la media dell'altezza dei genitori

Questi 2 grafici fa un rapporto tra le altezze dei figli, uno dei figli maschi e uno delle figlie femmine, e tra la media dei genitori

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Funzione per convertire pollici in centimetri
def inches_to_cm(inches):
    return inches * 2.54

# Leggi il file Galton.txt
data = pd.read_csv('Galton.txt', sep='\t')

# Converti le altezze da pollici a centimetri
data['Father_cm'] = inches_to_cm(data['Father'])
data['Mother_cm'] = inches_to_cm(data['Mother'])
data['Height_cm'] = inches_to_cm(data['Height'])

# Calcola l'altezza media dei genitori
data['Average_Parent_Height'] = (data['Father_cm'] + data['Mother_cm']) / 2

# Filtra i dati per maschi e femmine
data_male = data[data['Gender'] == 'M']
data_female = data[data['Gender'] == 'F']

# Funzione per creare il grafico di regressione
def plot_regression(data, gender):
    X = data[['Average_Parent_Height']]
    y = data['Height_cm']
    
    # Crea il modello di regressione lineare
    model = LinearRegression()
    model.fit(X, y)
    
    # Predici le altezze
    predictions = model.predict(X)
    
    # Crea il grafico
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Dati reali')
    plt.plot(X, predictions, color='red', linewidth=2, label='Regressione lineare')
    plt.title(f'Regressione Lineare dell\'Altezza dei {gender}')
    plt.xlabel('Altezza Media dei Genitori (cm)')
    plt.ylabel('Altezza dei Figli (cm)')
    plt.legend()
    plt.grid()
    plt.show()

# Crea i grafici per maschi e femmine
plot_regression(data_male, 'Maschi')
plot_regression(data_female, 'Femmine')
```

In base a questo grafico capiamo che l'altezza figli maschi è in media sempre maggiore rispetto a le figlie, i ragazzi arrivano quasi a 190cm, invece le ragazze arrivano quasi a 175cm

## Grafici lineare tra l'altezza dei figli e la media dell'altezza dei genitori

Questi 2 grafici fa un rapporto tra le altezze dei figli e tra la media dei genitori, uno con l'altezza del padre e uno con l'altezza della madre

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Funzione per convertire pollici in centimetri
def inches_to_cm(inches):
    return inches * 2.54

# Leggi il file Galton.txt
data = pd.read_csv('Galton.txt', sep='\t')

# Converti le altezze da pollici a centimetri
data['Father_cm'] = inches_to_cm(data['Father'])
data['Mother_cm'] = inches_to_cm(data['Mother'])
data['Height_cm'] = inches_to_cm(data['Height'])

# Funzione per creare il grafico di regressione
def plot_regression(X, y, title, xlabel):
    # Crea il modello di regressione lineare
    model = LinearRegression()
    model.fit(X, y)
    
    # Predici le altezze
    predictions = model.predict(X)
    
    # Crea il grafico
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Dati reali')
    plt.plot(X, predictions, color='red', linewidth=2, label='Regressione lineare')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Altezza dei Figli (cm)')
    plt.legend()
    plt.grid()
    plt.show()

# Grafico con l'altezza del padre
X_father = data[['Father_cm']]
y = data['Height_cm']
plot_regression(X_father, y, 'Regressione Lineare dell\'Altezza dei Figli (Padre)', 'Altezza del Padre (cm)')

# Grafico con l'altezza della madre
X_mother = data[['Mother_cm']]
plot_regression(X_mother, y, 'Regressione Lineare dell\'Altezza dei Figli (Madre)', 'Altezza della Madre (cm)')
```

In base a questi grafici siamo in grado di capire che l'altezza del padre, essendo mediamente più alto, è più impattante sull'altezza dei figli, rispetto all'altezza della madre
