```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leggi il dataset
df = pd.read_csv('gomma_data.csv')

# Seleziona solo le colonne numeriche di interesse
columns_of_interest = [
    'GROSS_WEIGHT_KG',
    'NET_WEIGHT_KG',
    'VOLUME_M3',
    'DECLARED_DISTANCE_KM',
    'DELIVERY_TIME_HH',
    'WDAY'
]

# Calcola la matrice di correlazione
correlation_matrix = df[columns_of_interest].corr()

# Crea una figura più grande per migliore leggibilità
plt.figure(figsize=(12, 10))

# Crea il correlogramma usando seaborn
sns.heatmap(correlation_matrix, 
            annot=True,  # Mostra i valori numerici
            cmap='RdBu',  # Usa una colormap blu-rosso
            vmin=-1, vmax=1,  # Imposta i limiti della scala
            center=0,  # Centra la colormap sullo zero
            fmt='.2f')  # Mostra due decimali

# Ruota le etichette per migliore leggibilità
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Aggiungi un titolo
plt.title('Correlogramma delle variabili di trasporto', pad=20)

# Aggiusta il layout
plt.tight_layout()

plt.show()
```

```python
import pandas as pd

# Read the data
df = pd.read_csv('.csv')

# Create a groupby object based on the departure details
grouped = df.groupby(['SHIPPING_DATE','DEPARTURE_ZIPCODE']).agg({
    'GROSS_WEIGHT_KG': 'sum',
    'NET_WEIGHT_KG': 'sum',
    # Count the number of occurrences for each group
    'SHIPPING_DATE': 'count'
}).rename(columns={'SHIPPING_DATE': 'count'})

# Filter for groups with more than 2 occurrences
filtered_groups = grouped[grouped['count'] > 2]

# Create the bus_cum list with tuples of (gross_weight, net_weight)
bus_cum = list(zip(filtered_groups['GROSS_WEIGHT_KG'], 
                  filtered_groups['NET_WEIGHT_KG']))

# Print the results
print("\nGroups with more than 2 identical departures:")
print(filtered_groups)

print("\nbus_cum list (gross_weight, net_weight):")
print(bus_cum)

# Print some summary statistics
print("\nSummary:")
print(f"Total number of groups found: {len(bus_cum)}")
if len(bus_cum) > 0:
    print(f"Maximum combined gross weight: {max(x[0] for x in bus_cum):.2f} kg")
    print(f"Maximum combined net weight: {max(x[1] for x in bus_cum):.2f} kg")
    print(f"Average combined gross weight: {sum(x[0] for x in bus_cum)/len(bus_cum):.2f} kg")
    print(f"Average combined net weight: {sum(x[1] for x in bus_cum)/len(bus_cum):.2f} kg")
```

```python
import pandas as pd

# Read the data
df = pd.read_csv('')

# Create a groupby object based on shipping date and departure zipcode
grouped = df.groupby(['SHIPPING_DATE', 'DEPARTURE_ZIPCODE']).agg({
    'GROSS_WEIGHT_KG': 'sum',
    'NET_WEIGHT_KG': 'sum',
    'SHIPPING_DATE': 'count'  # Count occurrences
}).rename(columns={'SHIPPING_DATE': 'count'})

# Filter for groups with more than 2 occurrences
filtered_groups = grouped[grouped['count'] > 2]

# Create bus_cum list with the aggregated weights
bus_cum = list(zip(filtered_groups['GROSS_WEIGHT_KG'], 
                  filtered_groups['NET_WEIGHT_KG']))

# Reset index to get SHIPPING_DATE and DEPARTURE_ZIPCODE as columns
filtered_groups = filtered_groups.reset_index()

# Create a new dataframe with required features
result_df = pd.DataFrame({
    'ID': range(len(filtered_groups)),
    'SHIPPING_DATE': filtered_groups['SHIPPING_DATE'],
    'DEPARTURE_ZIPCODE': filtered_groups['DEPARTURE_ZIPCODE'],
    'GROSS_WEIGHT_KG': filtered_groups['GROSS_WEIGHT_KG'],
    'NET_WEIGHT_KG': filtered_groups['NET_WEIGHT_KG']
})

# Save to CSV
result_df.to_csv('aggregated_shipments.csv', index=False)

# Print results
print("bus_cum list (gross_weight, net_weight):")
print(bus_cum)

print("\nFirst few rows of the new CSV:")
print(result_df.head())

print(f"\nTotal number of aggregated groups: {len(bus_cum)}")
print("\nNew CSV 'aggregated_shipments.csv' has been created.")
```
