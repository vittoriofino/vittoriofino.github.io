```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leggi il dataset
df = pd.read_csv('gomma_data.csv')

# Seleziona solo le colonne numeriche di interesse
columns_of_interest = [
    'GROSS_WEIGHT_KG',
    'NET_WEIGHT_KG',
    'VOLUME_M3',
    'DECLARED_DISTANCE_KM',
    'DELIVERY_TIME_HH',
    'WDAY'
]

# Calcola la matrice di correlazione
correlation_matrix = df[columns_of_interest].corr()

# Crea una figura pi grande per migliore leggibilit
plt.figure(figsize=(12, 10))

# Crea il correlogramma usando seaborn
sns.heatmap(correlation_matrix, 
            annot=True,  # Mostra i valori numerici
            cmap='RdBu',  # Usa una colormap blu-rosso
            vmin=-1, vmax=1,  # Imposta i limiti della scala
            center=0,  # Centra la colormap sullo zero
            fmt='.2f')  # Mostra due decimali

# Ruota le etichette per migliore leggibilit
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Aggiungi un titolo
plt.title('Correlogramma delle variabili di trasporto', pad=20)

# Aggiusta il layout
plt.tight_layout()

plt.show()
```
